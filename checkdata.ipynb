```python
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd

pairs = pd.read_csv("C:\Rashed\8119\Mid Term\scripts\data\sr\pairs_train.csv")
row = pairs.sample(1, random_state=0).iloc[0]
lr = Image.open(row.lr_path).convert("RGB")
hr = Image.open(row.hr_path).convert("RGB")

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.title("LR 3232"); plt.imshow(lr); plt.axis("off")
plt.subplot(1,2,2); plt.title("HR 128128"); plt.imshow(hr); plt.axis("off")
plt.tight_layout(); plt.show()
```
